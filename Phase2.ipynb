# Now Merge all board files and Clean Data

```python
All_boards_df = pd.merge(final_banu,final_adb , on='Roll No', how='outer')
```

```python
All_boards_df=pd.merge(final_bise_h,final_bise_mardan, on='Roll No', how='outer')
```

```python
All_boards_df = All_boards_df.dropna(subset=['GROUP_NAME'])
```
